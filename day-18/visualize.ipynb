```python
from matplotlib import pyplot as plt
import numpy as np

data = { tuple(map(int,l.split(','))) for l in open('input.txt')}
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

x = [d[0] for d in data]
y = [d[1] for d in data]
z = [d[2] for d in data]

ax.scatter3D(x, y, z, c=z, cmap='Reds');

plt.show()
```

```python
cubes = {tuple(map(int,l.split(','))) for l in open('input.txt')}
shifts = lambda x,y,z: {(x+1,y,z),(x-1,y,z),(x,y+1,z),(x,y-1,z),(x,y,z+1),(x,y,z-1)}

visited = set()
queue = [(-1,-1,-1)]

while len(queue) > 0:
    current = queue.pop()
    queue += [s for s in (shifts(*current) - cubes - visited) if all(-1<=c<=25 for c in s)]
    visited |= {current}

result = sum((s in visited) for c in cubes for s in shifts(*c))

print(result)
```
